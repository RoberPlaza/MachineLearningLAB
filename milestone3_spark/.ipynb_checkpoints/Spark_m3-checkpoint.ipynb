```python
filteredDF= df.filter((df.rowID % 10)== 0)
filteredDF.count()
```

```python
workingDF= filteredDF.drop('rain_accumulation').drop('rain_duration').drop('hpwren_timestamp')
```

```python
before = workingDF.count()
workingDF= workingDF.na.drop()
after=workingDF.count()
before -after
```

```python
Ahora vamos a escalar los datos el normalizado vaya
```

```python
from pyspark.ml.feature import VectorAssembler
featuredUsed=['air_pressure','air_temp','avg_wind_direction','avg_wind_speed','max_wind_direction','max_wind_speed','relative_humidity']
assembler = VectorAssembler(inputCols=featuredUsed, outputCol="features_unscaled")
assembled = assembler.transform(workingDF)
```

```python
from pyspark.ml.feature import StandardScaler

scaler= StandardScaler(inputCol="features_unscaled", outputCol="features",withStd=True, withMean=True)
scalerModel=scaler.fit(assembled)
scaledData=scalerModel.transform(assembled)
```

```python
scaledData = scaledData.select("features","rowID")
elbowset= scaledData.filter((scaledData.rowID % 3)==0).select("features")
elbowset.persist()
```

```python
import utils

clusters= range(2,31)
wsseList = utils.elbow(elbowset,clusters)
```

```python
utils.elbow_plot(wsseList,clusters)
```
